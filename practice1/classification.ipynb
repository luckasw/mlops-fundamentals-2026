```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_parquet("clean_tripdata.parquet")
```

```python
df['tipped'] = (df['tip_amount'] > 0).astype(int)
```

```python
df["tipped"].value_counts()
```

```python
initial_features = ['trip_distance', 'fare_amount', 'pickup_hour']
X = df[initial_features]
y = df['tipped']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python
initial_features = ['trip_distance', 'fare_amount', 'pickup_hour', "PULocationID"]
X = df[initial_features]
y = df['tipped']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

The accuracy and precicison of both models is about 55%. Recall and F1 score hover around 60%. 

Adding pickuplocation improved the metrics somewhat. Although the improvement is again very small.
